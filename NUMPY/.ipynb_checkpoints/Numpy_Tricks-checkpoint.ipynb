# Set functions

```python
m = np.array([1,2,3,4,5])
n = np.array([3,4,5,6,7])
```

```python
np.union1d(m,n)
```

```python
np.intersect1d(m,n)
```

```python
np.setdiff1d(m,n)
```

```python
np.setxor1d(m,n)
```

```python
np.in1d(m,n)
```
