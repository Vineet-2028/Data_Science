```python
import numpy as np
import pandas as pd
```

# Series from lists


```python
# string
country = ['India','China','USA','France','Russia']
pd.Series(country)
```

```python
# integers
runs = [13,24,56,78,44]
pd.Series(runs,dtype='int32')
```

```python
# custom index
marks = [67,56,89,99]
subjects = ['Maths','English','Science','Hindi']
pd.Series(marks,index=subjects)
```

```python
# setting a name
pd.Series(marks,index=subjects,name='Vineet Chauhan ke marks')
```

# Series from Dict

```python
marks = {
    'Maths':67,
    "English":57,
    'Science':89,
    "Hindi":100
}
marks_series = pd.Series(marks,name='Vineet Chauhan ke marks')
marks_series
```

# Boolean Indexing on Series

```python
v = pd.read_csv('kohli_ipl.csv',index_col='match_no')
s = v .squeeze("columns")   
print(s)
```

```python
# Find number of ducks
v[v  == 0]
```

```python
# Count numer of day when I had more than 200 subs a day
df[df > 200].size
```

```python
# FInd actors who have done more than 20 movies
num_movies = bm.value_counts()
num_movies[num_movies > 20]
```
