# Filtering DataFrame

```python
ipl
```

```python
# find all the final winner
mask = ipl['MatchNumber'] == 'Final'
new_df = ipl[mask]
new_df[['Season','WinningTeam']]
```

```python
# how many super over finishes have occured
ipl[ipl['SuperOver'] == 'Y'].shape[0]
```

```python
# how many matches has rcb won in kolkata
ipl[ipl['WinningTeam'] == 'Royal Challengers Bangalore'].shape[0]
```

```python
ipl[(ipl['City'] == 'Kolkata') & (ipl['WinningTeam'] == 'Royal Challengers Bangalore')].shape[0]
```

```python
# toss winner is match winner in percentage
(ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0]/ipl.shape[0])*100
```

```python
# movies with rating higher than 8 and vote  > 10000
movies[(movies['imdb_rating'] > 8.5) & (movies['imdb_votes'] > 10000)].shape[0]
```
